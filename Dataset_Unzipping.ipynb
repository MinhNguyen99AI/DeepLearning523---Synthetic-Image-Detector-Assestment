# Displaying contents and counts of images

```python
import os
initial_count = 0
dir = "./progan_train"
for path in os.listdir(dir):
    dir_2 = os.path.join(dir, path)
    print(initial_count)
    print(dir_2)
    for path_2 in os.listdir(dir_2):
        dir_3 = os.path.join(dir_2, path_2)
        print(initial_count)
        print(dir_3)
        for path_3 in os.listdir(dir_3):
            #print(os.path.join(dir_3, path_3))
            #if os.path.isfile(os.path.join(dir_3, path_3)):
            initial_count += 1
print(initial_count)
```

```python
initial_count = 0
dir = "./CNN_synth_testset"
for path in os.listdir(dir):
    dir_2 = os.path.join(dir, path)
    print(initial_count)
    print(dir_2)
    for path_2 in os.listdir(dir_2):
        dir_3 = os.path.join(dir_2, path_2)
        print(initial_count)
        print(dir_3)
        for path_3 in os.listdir(dir_3):
            #print(os.path.join(dir_3, path_3))
            dir_4 = os.path.join(dir_3, path_3)
            if not os.path.isfile(dir_4):
                print(initial_count)
                print(dir_4)
                for path_4 in os.listdir(dir_4):
                    initial_count += 1
            else:
                initial_count += 1
print(initial_count)
```
